```python
# O tipo de cartÃ£o estÃ¡ relacionado com o churn?
# Limite de crÃ©dito e churn estÃ£o correlacionados?
df_churn = pd.read_csv("./BankChurners.csv")
def grafico_comparativo(cartao,perc_churn):
    plt.bar(cartao, perc_churn)
    plt.title('Churn por Tipo cartao')
    plt.xlabel('Cartao')
    plt.ylabel('Percentual Churn')
```

```python
df_churn.replace({'Education_Level':{'Unknown': np.nan},
                  'Marital_Status':{'Unknown': np.nan},
                  'Income_Category':{'Unknown': np.nan}
                }, inplace = True)

df_churn['media_credit_card'] = df_churn.groupby(['Card_Category'])[['Credit_Limit']].transform('mean')
df_churn.head(5)
```

```python
cartao = []
perc_churn = []

for card in df_churn.Card_Category.unique():
    print(card)
    for churn in df_churn.Attrition_Flag.unique():
        clientes_card_churn = df_churn[
                                        (df_churn['Card_Category']==card) &
                                        (df_churn['Attrition_Flag']==churn)
                                        ].Attrition_Flag.count()
        
        clientes_card = df_churn[df_churn['Card_Category']==card].Attrition_Flag.count()
        clientes_card_churnc = df_churn[
                                        (df_churn['Card_Category']==card) &
                                        (df_churn['Attrition_Flag']==churn)
                                        ].Credit_Limit.mean()
        
        clientes_cardc = df_churn[df_churn['Card_Category']==card].Credit_Limit.mean()
        
        idade_cardc = df_churn[
                                        (df_churn['Card_Category']==card) & 
                                        (df_churn['Attrition_Flag']==churn)
                                        ].Customer_Age.mean()
        
        idade_card= df_churn[df_churn['Card_Category']==card].Customer_Age.mean()
        
        if churn == "Attrited Customer":
            cartao.append(card)
            perc_churn.append((clientes_card_churn/ clientes_card)*100)
        
        print(f'Clientes {churn} - sÃ£o {clientes_card_churn}',
              f',{(clientes_card_churn/ clientes_card)*100 :,.2f}% do grupo. Media de credito {clientes_card_churnc/1000 :,.0f}K. Media idade {idade_cardc :,.2f}')
    
    print(f'A media de credito do cartao e {clientes_cardc :,.2f}. Com media de idade de {idade_card :,.1f} anos')
    print()
    
print(cartao,perc_churn)
```
