```python
# O tipo de cartão está relacionado com o churn?
# Limite de crédito e churn estão correlacionados?
df_churn = pd.read_csv("./BankChurners.csv")
def grafico_comparativo(cartao,perc_churn):
    plt.bar(cartao, perc_churn)
    plt.title('Churn por Tipo cartao')
    plt.xlabel('Cartao')
    plt.ylabel('Percentual Churn')
```

```python
df_churn.replace({'Education_Level':{'Unknown': np.nan},
                  'Marital_Status':{'Unknown': np.nan},
                  'Income_Category':{'Unknown': np.nan}
                }, inplace = True)

df_churn['media_credit_card'] = df_churn.groupby(['Card_Category'])[['Credit_Limit']].transform('mean')
df_churn.head(5)
```

```python
cartao = []
perc_churn = []

for card in df_churn.Card_Category.unique():
    print(card)
    for churn in df_churn.Attrition_Flag.unique():
        clientes_card_churn = df_churn[
                                        (df_churn['Card_Category']==card) &
                                        (df_churn['Attrition_Flag']==churn)
                                        ].Attrition_Flag.count()
        
        clientes_card = df_churn[df_churn['Card_Category']==card].Attrition_Flag.count()
        clientes_card_churnc = df_churn[
                                        (df_churn['Card_Category']==card) &
                                        (df_churn['Attrition_Flag']==churn)
                                        ].Credit_Limit.mean()
        
        clientes_cardc = df_churn[df_churn['Card_Category']==card].Credit_Limit.mean()
        
        idade_cardc = df_churn[
                                        (df_churn['Card_Category']==card) & 
                                        (df_churn['Attrition_Flag']==churn)
                                        ].Customer_Age.mean()
        
        idade_card= df_churn[df_churn['Card_Category']==card].Customer_Age.mean()
        
        if churn == "Attrited Customer":
            cartao.append(card)
            perc_churn.append((clientes_card_churn/ clientes_card)*100)
        
        print(f'Clientes {churn} - são {clientes_card_churn}',
              f',{(clientes_card_churn/ clientes_card)*100 :,.2f}% do grupo. Media de credito {clientes_card_churnc/1000 :,.0f}K. Media idade {idade_cardc :,.2f}')
    
    print(f'A media de credito do cartao e {clientes_cardc :,.2f}. Com media de idade de {idade_card :,.1f} anos')
    print()
    
print(cartao,perc_churn)
```

```python
df2=pd.get_dummies(df, columns=['Acc_status'], drop_first=False)
```

```python
# Agrupando dados utilizando o tempo total como cliente, em segmentos de igual frequencia,
#é possivel verificar que nao existe uma relação entre o tempo de conta e a quantidade de churn
df2.groupby(pd.qcut(df2.Months_on_book, 3))[['Acc_status_Attrited Customer']].sum()/df2.groupby(pd.qcut(df2.Months_on_book, 3))[['Acc_status_Attrited Customer']].count()
```

```python
# Agrupando os dados, também em quartis de igual freqüencia,
#é possivel verificar que exite uma maior concentraçao de churn em função do ratio de
# uso do cartão em relaçao ao total de limite disponível, levando a crer que pessoas que utlizam menos
# ou que dependem menos do cartão,
# tem uma probabilidade maior de cancelamento.
# É necessária uma avaliação mais completa, 
# mas é possivel afirmar a importancia desse indicador através do gráfico 
```

```python
df2.groupby(pd.qcut(df2.Avg_Utilization_Ratio, 3))[['Acc_status_Attrited Customer']].sum()/df2.groupby(pd.qcut(df2.Avg_Utilization_Ratio, 3))[['Acc_status_Attrited Customer']].count()
```

```python
# utilização média do cartão por status 
avg_use = df.groupby("Acc_status")[['Avg_Utilization_Ratio']].mean()
avg_use
```

```python
# Adicionalmente é possivel avaiar que mesmo combinadas as duas condições,
# o grau de utilização ainda é preponderante em relação ao tempo como cliente.
```

```python
df2.groupby([pd.qcut(df2.Months_on_book, 3),pd.qcut(df2.Avg_Utilization_Ratio, 3)])[['Acc_status_Attrited Customer']].sum()/df2.groupby([pd.qcut(df2.Months_on_book, 3),pd.qcut(df2.Avg_Utilization_Ratio, 3)])[['Acc_status_Attrited Customer']].count()
```
